```python
import pandas as pd
```

خواندن دیتاست

```python
df = pd.read_csv('diabetes.csv')
df
```

ایجاد مدل طبقه بند کا همسایه

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=5)
```

```python
y = df['Outcome']
X = df.drop('Outcome', axis=1)
X
```

```python
model.fit(X, y)
```

استفاده از مدل

```python
model.predict([
    [1, 86, 57, 37, 100, 26, 0.2, 35],
    [4, 86, 57, 37, 100, 41, 0.2, 27],
              ])
```

```python
yhat = model.predict(X)
yhat[:10]
```

```python
y.values[:10]
```

ارزیابی کیفیت مدل

```python
import numpy as np
(len(yhat)-np.sum(abs(y.values-yhat)))/len(yhat)
```

```python
np.sum(y.values==yhat)/len(yhat)
```

چرخ را از اول اختراع نکنیم. استفاده از امکانات کتابخانه

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y, yhat)
```

معیار اکیوریسی برای داده نامتعادل خوب نیست. دلیلش در کلاس بحث شد

معیارهای ارزیابی بهتری برای این موارد وجود دارد

تئوری معیارهای ارزیابی مختلف در کلاس تدریس شد

```python
from sklearn.metrics import  f1_score, classification_report, confusion_matrix
```

```python
f1_score(y, yhat)
```

```python
print(classification_report(y, yhat))
```

```python
confusion_matrix(y, yhat)
```

ارزیابی مدل با همان داده ای که برای ساخت مدل استفاده شده کار صحیحی نیست

دلیلش در کلاس بحث شد

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
yhat = model.predict(X_test)
accuracy_score(y_test, yhat)
```

```python
print(classification_report(y_test, yhat))
```

هر الگوریتم یادگیری معمولا یک سری پارامتر دارد

مثلا در الگوریتم کا همسایه عدد کا چند باشد؟

```python
model = KNeighborsClassifier(n_neighbors=10)
model.fit(X_train, y_train)
yhat = model.predict(X_test)
accuracy_score(y_test, yhat)
```

```python
print(classification_report(y_test, yhat))
```

با کا بزرگتر کیفیت بهتر شد
