```python
from collections import defaultdict
```

### read file

```python
file = open('super_market_sales.csv')
lines = file.readlines()
```

### show head

```python
lines[1]
```

split file 

```python
data = []
for line in lines[1:]:
    col = line.split(',')
    data.append((col[1], col[2],col[15]))
data
```

### using defaultdict to add all data in one list and creat new dic


```python
#for cod,name,value in data:
d=defaultdict(list)
for c,n,v in data:
    d[c,n].append(float(v))
print(d)
i=0.0
m=dict(d)

```

### sum 

```python
l1=m['A', 'Yangon']
l2=m['B', 'Mandalay']
l3=m['C', 'Naypyitaw']
x1=sum(l1)
x2=sum(l2)
x3=sum(l3)
print(l4)
m['A', 'Yangon']=x1
m['B', 'Mandalay']=x2
m['C', 'Naypyitaw']=x3
```
