# Real-world clustering example

```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
odf = pd.read_csv('players_21.csv')
odf
```

```python
odf.columns.values
```

```python
df = odf[['short_name', 'age', 'height_cm', 'weight_kg', 'league_rank',
    'value_eur', 'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic'
   ]].dropna().set_index('short_name')
df
```

ساده ترین حالت
وقتی که همه ستون ها عددی باشند

```python
model = KMeans(n_clusters=10)
model.fit(df.values)
model.cluster_centers_
```

```python
df['cluster'] = model.labels_
df
```

```python
df[df.cluster==9]
```

```python
df[df.cluster==7]
```

# Preprocessing: scale data to (0, 1)
دلیل نیاز به این کار
برای حل مشکل زیر
محدوده خصوصیات خیلی با هم متفاوتند. مثلا وزن و قیمت
باعث میشه مثلا اثر قیمت در فاصله اقلیدسی اثر وزن را از بین ببرد

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
scaled
```

```python
model = KMeans(n_clusters=10)
model.fit(scaled)
model.cluster_centers_
```

```python
df['cluster'] = model.labels_
df
```

```python
df[df.cluster==4]
```

حالت بعد

وقتی بخواهیم خصوصیات غیر عددی را هم استفاده کنیم

مثلا پای ارجح

راه حل

به هر مقدار غیر عددی یک عدد اختصاص دهیم

```python
df = odf[['short_name', 'preferred_foot', 'age', 'height_cm', 'weight_kg', 'league_rank',
    'value_eur', 'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic'
   ]].dropna().set_index('short_name')
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
df['preferred_foot'] = enc.fit_transform(df['preferred_foot'])
df
```

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
model = KMeans(n_clusters=10)
model.fit(scaled)
model.cluster_centers_
df['cluster'] = model.labels_
df
```
