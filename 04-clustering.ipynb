```python
import pandas as pd
from sklearn.cluster import KMeans, MeanShift
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fake.csv')
df
```

```python
df.plot.scatter('x', 'y')
```

```python
model = KMeans(n_clusters=3)
model.fit(df)
```

```python
c = model.cluster_centers_.T
c
```

```python
df.plot.scatter('x', 'y')
plt.scatter(c[0], c[1], color='red')
```

```python
model = KMeans(n_clusters=2)
model.fit(df)
c = model.cluster_centers_.T
df['label'] = model.labels_
ax = df[df.label==0].plot.scatter('x', 'y', color='blue')
df[df.label==1].plot.scatter('x', 'y', color='green', ax=ax)
plt.scatter(c[0], c[1], color='red')
plt.show()
```

```python
model = MeanShift()
model.fit(df)
c = model.cluster_centers_.T
df['label'] = model.labels_
ax = df[df.label==0].plot.scatter('x', 'y', color='blue')
df[df.label==1].plot.scatter('x', 'y', color='green', ax=ax)
df[df.label==2].plot.scatter('x', 'y', color='orange', ax=ax)
plt.scatter(c[0], c[1], color='red')
plt.show()
```
